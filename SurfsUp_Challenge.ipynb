```python
# Dependencies
import numpy as np

# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
```

```python
import datetime as dt
import pandas as pd
```

```python
%matplotlib inline
from matplotlib import style
style.use('fivethirtyeight')
import matplotlib.pyplot as plt
```

```python
engine = create_engine("sqlite:///hawaii.sqlite")

# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)

# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

## D1: Determine the Summary Statistics for June

```python
# 1. Import the sqlalchemy extract function.
from sqlalchemy import extract

# 2. Write a query that filters the Measurement table to retrieve the temperatures for the month of June. 
june_results = session.query(Measurement.date, Measurement.tobs).\
    filter(extract('month', Measurement.date)==6)
```

```python
#  3. Convert the June temperatures to a list.
june_results = session.query(Measurement.date, Measurement.tobs).\
    filter(extract('month', Measurement.date)==6).all()
print(june_results)
```

```python
# 4. Create a DataFrame from the list of temperatures for the month of June. 
june_df = pd.DataFrame(june_results, columns=['date', 'June Temps'])
```

```python
# 5. Calculate and print out the summary statistics for the June temperature DataFrame.
print(june_df.describe().to_string(index=True))
```

```python
#Df Format
june_df.describe()
```

## D2: Determine the Summary Statistics for December

```python
# 6. Write a query that filters the Measurement table to retrieve the temperatures for the month of December.
dec_results = session.query(Measurement.date, Measurement.tobs).\
    filter(extract('month', Measurement.date)==12)
print(dec_results)
```

```python
# 7. Convert the December temperatures to a list.
dec_results = session.query(Measurement.date, Measurement.tobs).\
    filter(extract('month', Measurement.date)==12).all()
print(dec_results)
```

```python
# 8. Create a DataFrame from the list of temperatures for the month of December. 
dec_df = pd.DataFrame(dec_results, columns=['date', 'Dec Temps'])
dec_df
```

```python
# 9. Calculate and print out the summary statistics for the Decemeber temperature DataFrame.
print(dec_df.describe().to_string(index=True))
```

```python
#Df Format
dec_df.describe()
```

```python
# Plot June results to visualize summary statistics
june_df.plot.hist(bins=20)
plt.tight_layout()
```

```python
# Plot December results to visualize summary statistics
dec_df.plot.hist(bins=20)
plt.tight_layout()
```

```python
#Additional precipitation information in June and Dec
session.query(Measurement.tobs).filter(extract('month',Measurement.date) == 6).all()
session.query(Measurement.tobs).filter(extract('month',Measurement.date) == 12).all()


Jun_precipitation = session.query(Measurement.prcp).filter(extract('month', Measurement.date) == 6).all()
Dec_precipitation = session.query(Measurement.prcp).filter(extract('month', Measurement.date) == 12).all()

Jun_prcp_df = pd.DataFrame(Jun_precipitation, columns=['June Precipitation'])
Dec_prcp_df = pd.DataFrame(Dec_precipitation, columns= ['December Precipitation'])

Jun_prcp_df.describe()
```
